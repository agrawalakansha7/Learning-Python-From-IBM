```python
#Copy and Clone List

A = ["akansha agrawal", 10, 1.2]
B = A
print('A:', A)
print('B:', B)
```

```python
# Examine the copy by reference

print('B[0] before changing A[0]:', B[0])
A[0] = "banana"
print('B[0] after changing A[0]:', B[0])
```

```python
# Clone (clone by value) the list A

B = A[:]
B
```

```python
print('B[0]:', B[0])
A[0] = "akansha agrawal"
print('B[0]:', B[0])
```
